```python
covid_df.drop(['Sno','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1,inplace=True)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
covid_df['Active_Cases']= covid_df['Confirmed']-(covid_df['Cured'])+covid_df['Deaths']
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values=["Confirmed", "Deaths", "Cured"], index="State/UnionTerritory", aggfunc=max)
```

```python
statewise["Recovery Rate"]= statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Morality Rate"]= statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by= "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
 # Top 10 active cases
top_active = covid_df.groupby(by='State/UnionTerritory') \
    .max()[['Active_Cases', 'Date']] \
    .sort_values(by='Active_Cases', ascending=False) \
    .reset_index()

top_active.head(10)
```

```python
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with Most Active Cases in India", size=25)

# Use a color palette (like "viridis", "Set2", or "pastel", etc.)
ax = sns.barplot(
    data=top_active.iloc[:10],
    y='Active_Cases',
    x='State/UnionTerritory',
    linewidth=2,
    edgecolor='black',
    palette='Set2'  # <== Add this line to change colors
)

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.xticks(rotation=45)  # Optional: Rotate x-axis labels for better visibility
plt.show()

```

```python
# Top states with highest deaths
top_deaths = covid_df.groupby(by='State/UnionTerritory') \
    .max()[['Deaths', 'Date']] \
    .sort_values(by='Deaths', ascending=False) \
    .reset_index()

fig = plt.figure(figsize=(18, 5))
plt.title("Top 10 States with Most Deaths", size=25)

ax = sns.barplot(
    data=top_deaths.iloc[:10],  # 10 instead of 12 for top 10
    y="Deaths",
    x="State/UnionTerritory",
    linewidth=2,
    edgecolor="black",
    palette="Reds"  # Optional: adds color variation
)

plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.xticks(rotation=45)  # Optional: improves readability
plt.show()
```

```python
# Growth Trends for top 5 states
fig = plt.figure(figsize=(12, 6))
ax = sns.lineplot(
    data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])],
    x='Date',
    y='Active_Cases',
    hue='State/UnionTerritory'
)

plt.title("Growth Trend of Active Cases in Selected States", size=18)
plt.xlabel("Date")
plt.ylabel("Active Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Remove rows where state = India
vaccine = vaccine_df[vaccine_df.State!='India']
print(vaccine)
```

```python
# Most vaccinated states (Top 5)
max_vac = vaccine.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending=False).head(5)
max_vac
```

```python
fig = plt.figure(figsize=(10, 5))
plt.title("Top 5 Vaccinated States in India", size=20)
ax = sns.barplot(data=max_vac.iloc[:10], x=max_vac.index, y=max_vac.Total, linewidth=2, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Least vaccinated states (Bottom 5)
min_vac = vaccine.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total')
min_vac = min_vac.sort_values('Total', ascending=True).head(5)
min_vac

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(10, 5))
plt.title("Bottom 5 Vaccinated States in India", size=20)
ax = sns.barplot(data=min_vac, x=min_vac.index, y=min_vac.Total, linewidth=2, edgecolor='black', palette='Reds')
plt.xlabel("States")
plt.ylabel("Total Vaccinations")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a risk level column
def risk_level(row):
    if row['Active_Cases'] > 100000:
        return 'High'
    elif row['Active_Cases'] > 20000:
        return 'Moderate'
    else:
        return 'Low'

covid_df['Risk_Level'] = covid_df.apply(risk_level, axis=1)

# Feature set
features = covid_df[['Confirmed', 'Deaths', 'Cured']]
labels = covid_df['Risk_Level']

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
print(classification_report(y_test, clf.predict(X_test)))
```
